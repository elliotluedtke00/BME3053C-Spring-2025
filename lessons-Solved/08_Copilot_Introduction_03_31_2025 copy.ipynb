# What is GitHub Copilot?

GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

````python
# Create a scatter plot using matplotlib with the following features:
# - Random data points
# - Custom colors based on point values
# - A trend line
# - Proper labels and title
# - A legend
# - Proper

# Let Copilot suggest the implementation
# Import required libraries
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np

# Set the style for better visualization
plt.style.use('default')  # Using default style instead of seaborn

# Generate random data
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 10, 50)
y = 2 * x + np.random.normal(0, 2, 50)

# Create the figure
plt.figure(figsize=(10, 6))

# Create scatter plot with color mapping
scatter = plt.scatter(x, y, 
                     c=y,                    # Color based on y values
                     cmap='viridis',         # Color map
                     alpha=0.6,              # Transparency
                     s=100,                  # Point size
                     label='Data Points')    # Label for legend

# Calculate and add trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), 
         color='red',
         linestyle='--', 
         linewidth=2,
         label=f'Trend Line (y = {z[0]:.2f}x + {z[1]:.2f})')

# Customize the plot
plt.title('Scatter Plot with Trend Line', 
         fontsize=14, 
         pad=20)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Point Values', fontsize=10)

# Add legend
plt.legend(fontsize=10)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
````

### ✏️ Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
import numpy as np
from scipy.ndimage import gaussian_filter

def process_medical_image(image):
    """
    Process a medical image by normalizing and applying a Gaussian filter.

    Parameters:
        image (numpy.ndarray): Input image as a numpy array.

    Returns:
        numpy.ndarray: Processed image.

    Raises:
        ValueError: If the input is not a numpy array or is empty.
    """
    if not isinstance(image, np.ndarray):
        raise ValueError("Input must be a numpy array.")
    if image.size == 0:
        raise ValueError("Input image cannot be empty.")

    # Normalize the image to the range [0, 1]
    normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Apply a Gaussian filter for smoothing
    filtered_image = gaussian_filter(normalized_image, sigma=1)

    return filtered_image
````


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    def __init__(self):
        """Initialize an empty dictionary to store patient records."""
        self.records = {}

    def add_record(self, patient_id, name, age, condition):
        """
        Add a new patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            name (str): Name of the patient.
            age (int): Age of the patient.
            condition (str): Medical condition of the patient.

        Raises:
            ValueError: If the data is invalid or patient_id already exists.
        """
        if patient_id in self.records:
            raise ValueError(f"Patient ID {patient_id} already exists.")
        if not isinstance(age, int) or age <= 0:
            raise ValueError("Age must be a positive integer.")
        if not name or not condition:
            raise ValueError("Name and condition cannot be empty.")

        self.records[patient_id] = {"name": name, "age": age, "condition": condition}

    def update_record(self, patient_id, **kwargs):
        """
        Update an existing patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            kwargs: Fields to update (e.g., name, age, condition).

        Raises:
            KeyError: If the patient_id does not exist.
        """
        if patient_id not in self.records:
            raise KeyError(f"Patient ID {patient_id} does not exist.")

        for key, value in kwargs.items():
            if key in self.records[patient_id]:
                self.records[patient_id][key] = value

    def get_record(self, patient_id):
        """
        Retrieve a patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.

        Returns:
            dict: Patient record.

        Raises:
            KeyError: If the patient_id does not exist.
        """
        if patient_id not in self.records:
            raise KeyError(f"Patient ID {patient_id} does not exist.")
        return self.records[patient_id]

    def average_age(self):
        """
        Calculate the average age of all patients.

        Returns:
            float: Average age of patients, or 0 if no records exist.
        """
        if not self.records:
            return 0
        total_age = sum(record["age"] for record in self.records.values())
        return total_age / len(self.records)

    def condition_statistics(self):
        """
        Calculate the count of patients for each condition.

        Returns:
            dict: A dictionary with conditions as keys and counts as values.
        """
        stats = {}
        for record in self.records.values():
            condition = record["condition"]
            stats[condition] = stats.get(condition, 0) + 1
        return stats
````

````python
# Create an instance of PatientRecords
records = PatientRecords()

# Add patient records
records.add_record("P001", "Alice", 30, "Diabetes")
records.add_record("P002", "Bob", 45, "Hypertension")
records.add_record("P003", "Charlie", 25, "Diabetes")

# Update a record
records.update_record("P001", age=31)

# Retrieve a record
print("Record for P001:", records.get_record("P001"))

# Perform statistical analysis
print("Average Age:", records.average_age())
print("Condition Statistics:", records.condition_statistics())
````

### ✏️ Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

def create_dashboard(data):
    """
    Create a data visualization dashboard.

    Parameters:
        data (pd.DataFrame): Input data for visualization.
    """
    def plot_data(column, plot_type):
        """
        Plot the selected column with the specified plot type.

        Parameters:
            column (str): Column to visualize.
            plot_type (str): Type of plot (e.g., 'line', 'bar', 'hist').
        """
        plt.figure(figsize=(10, 6))
        if plot_type == 'line':
            data[column].plot(kind='line', title=f"{column} - Line Plot")
        elif plot_type == 'bar':
            data[column].value_counts().plot(kind='bar', title=f"{column} - Bar Plot")
        elif plot_type == 'hist':
            data[column].plot(kind='hist', bins=10, title=f"{column} - Histogram")
        else:
            print("Unsupported plot type.")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.grid(True)
        plt.show()

    # Create interactive widgets
    column_selector = widgets.Dropdown(
        options=data.columns,
        description="Column:",
        continuous_update=False
    )
    plot_type_selector = widgets.Dropdown(
        options=['line', 'bar', 'hist'],
        description="Plot Type:",
        continuous_update=False
    )

    interact(plot_data, column=column_selector, plot_type=plot_type_selector)

# Example usage
if __name__ == "__main__":
    # Sample data
    data = pd.DataFrame({
        "Age": [25, 30, 35, 40, 45, 50, 55, 60],
        "Height": [160, 165, 170, 175, 180, 185, 190, 195],
        "Weight": [50, 55, 60, 65, 70, 75, 80, 85],
        "Condition": ["Healthy", "Diabetes", "Healthy", "Hypertension", "Healthy", "Diabetes", "Healthy", "Hypertension"]
    })

    create_dashboard(data)
````
